```python
# Тест 0
A0 = np.array([[0, 2, 3], [1, 2, 4], [4, 5, 6]], dtype='float64')
b0 = np.array([13, 17, 32], dtype='float64')

# Тест 1
A1 = np.array([[7, 1, 1], [1, 9, 1], [1, 1, 11]], dtype='float64')
b1 = np.array([9, 11, 13], dtype='float64')

# Тест 2
A2 = np.array([[-7, 1, 1], [1, -9, 1], [1, 1, -11]], dtype='float64')
b2 = np.array([-9, -11, -13], dtype='float64')

# Тест 3
A3 = np.array([[-7, 8, 9], [10, -9, 6], [9, 10, -11]], dtype='float64')
b3 = np.array([9, 11, 13], dtype='float64')

# Тест 4
A4 = np.array([[7, 6, 6], [6, 9, 6], [6, 6, 11]], dtype='float64')
b4 = np.array([9, 11, 13], dtype='float64')


all_A = [A0, A1, A2, A3, A4]
all_b = [b0, b1, b2, b3, b4]


# Пятый тест
def fifth_matr(n, eps):
    X = np.eye(n) # A
    Y = np.ones([n, n]) # +eps*N*Y
    for i in range(n):
        for j in range(n):
            if i < j:
                X[i, j] = -1
                Y[i, j] = -1

    A = matrix_sum(X, eps * 5 * Y)
    b = -1 *np.ones(n)
    b[-1] = 1
    return A, b
```

matrix function


```python
def matrix_sum(A, B):
    sum = np.zeros(A.shape)
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            sum[i, j] = A[i,j] + B[i,j]
    return sum

def mult_matrix(A, B):
    if len(B.shape) == 1:
        mult = np.zeros(A.shape[0])
        for i in range(mult.shape[0]):
            mult[i] = sum(A[i] * B)
        return mult
    else:
        mult = np.zeros((A.shape[0],  B.shape[1]))
        for i in range(mult.shape[0]):
            for j in range(mult.shape[1]):
                mult[i, j] = sum(A[i] * B[:, j])
        return mult
    
def matrix_T(A):
    T = np.zeros((A.shape[1], A.shape[0]))
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            T[i,j] = A[j,i]
    return T

def mysqrt(num, eps = 0.00001):
    w0 = num / 2 
    w1 = 0.5 * (w0 + num / w0)
    while abs(w1 - w0) > eps:
        w0 = w1
        w1 = 0.5 * (w0 + num / w0)
    return w1

def matrix_minor(A,i,j):
    matr = A.tolist()
    matr = matr[:i] + matr[i+1:]
    for p in range(2):
        matr[p] = matr[p][:j] + matr[p][j+1:]
    return np.array(matr)

def det(A, n):
    if n == 1:
        return A[0][0]
    if n == 2:
        return A[0][0] * A[1][1] - A[0][1] * A[1][0]
    res = 0
    for i in range(n):
        mnr = matrix_minor(A, 0, i)
        res += ((-1)**i) * A[0][i] * det(mnr, n-1)
    return res  
```

Norms


```python
def vector_norm_1(v):
    return sum(abs(v))

def vector_norm_2(v):
    return mysqrt(sum(v**2))

def vector_norm_inf(v): 
    return max(abs(v))

def vector_norm(v):
    return np.array([vector_norm_1(v), vector_norm_2(v), vector_norm_inf(v)])
    
    

def matr_norm_1(A):
    return max(sum(abs(A)))

def matr_norm_2(A):
    matrix = mult_matrix(matrix_T(A), A)
    return mysqrt(max(np.linalg.eigh(matrix)[0]))

def matr_norm_inf(A):
    return max(sum(abs(matrix_T(A))))

def matrix_norm(A):
    return np.array([matr_norm_1(A), matr_norm_2(A), matr_norm_inf(A)])
```

## Метод простой итерации

```python
def simple_iter(A, b, eps): 
    b = mult_matrix(matrix_T(A), b)
    A = mult_matrix(matrix_T(A), A)
    norm_A = matrix_norm(A)[1]
    mu = 1 / norm_A
    B = matrix_sum(np.eye(A.shape[0]), -mu * A)
    norm_B = matrix_norm(B)[1]

    
    x0 = mu * b
    x1 = mult_matrix(B, x0) + x0
    iterations = 1
    while vector_norm_2(mult_matrix(A, x1)-b) > eps:
        iterations += 1
        x0 = x1
        x1 = mult_matrix(B, x0) + (mu * b)
    return x1, iterations
```

```python
all_eps = [10**(-3), 10**(-4), 10**(-5)]
all_A = [A0, A1, A2, A3, A4]
all_b = [b0, b1, b2, b3, b4]


tests = [] 
py_solution = []
eps = []
my_solution = []
error = []
num_iter = []


for ind_test in range(5):
  for e in all_eps:
        tests.append(ind_test)
        python_solution = np.linalg.solve(all_A[ind_test], all_b[ind_test])
        py_solution.append(python_solution)
        eps.append(e)
        solution = simple_iter(all_A[ind_test], all_b[ind_test], e)
        
        if not isinstance(solution, str):
            my_solution.append(solution[0])
            error.append(abs(solution[0] - python_solution))
            num_iter.append(solution[1])
        else:
            my_solution.append(solution)
            error.append('-')
            num_iter.append('-')

data = {'test': tests,
        'Точное решение': py_solution,
        'Допустимая погрешность решения': eps,
        'Полученное решение': my_solution,
        'Абсолютная погрешность решения': error,
        'Количество итераций': num_iter}

pd.DataFrame(data)
```

## Метод Зейделя

```python
def Seidel(A, b, eps):
    
    def make_C_d(A, b):
        C = np.zeros(A.shape)
        d = np.zeros(b.shape)
        for i in range(A.shape[0]):
            d[i] = b[i] / A[i,i] 
            for j in range(A.shape[1]):
                if i == j:
                    C[i, j] = 0
                else:
                    C[i, j] = -A[i, j] / A[i,i] 
        return C, d


    def make_x1(C, d, x0):
        x1 = np.copy(d)
        for i in range(C.shape[0]):
            for j in range(C.shape[0]):
                if j > i: 
                    x1[i] += C[i,j] * x0[j]
                else:
                    x1[i] += C[i,j] * x1[j]
        return x1
    
    def diag_dominance(A):
        D = np.diag(np.abs(A))
        S = np.sum(np.abs(A), axis=1) - D 
        if np.all(D > S):
            return True
        else:
            return False
    

    

    if diag_dominance(A):
        C, d = make_C_d(A, b)
        x0 = np.copy(d)
        x1 = make_x1(C, d, x0)
        iterations = 1
        while vector_norm(mult_matrix(A, x1) - b)[0] > eps:
            iterations += 1
            x0 = x1
            x1 = make_x1(C, d, x0)
        return x1, iterations
    else: 
      b = mult_matrix(matrix_T(A), b)
      A = mult_matrix(matrix_T(A), A)
      C, d = make_C_d(A, b)
      x0 = np.copy(d)
      x1 = make_x1(C, d, x0)
      iterations = 1
      while vector_norm(mult_matrix(A, x1) - b)[0] > eps:
        iterations += 1
        x0 = x1
        x1 = make_x1(C, d, x0)
      return x1, iterations
```

```python
all_eps = [10**(-3), 10**(-4), 10**(-5)]
all_A = [A0, A1, A2, A3, A4]
all_b = [b0, b1, b2, b3, b4]


tests = [] 
py_solution = []
eps = []
my_solution = []
error = []
num_iter = []

for ind_test in range(5):
        tests.append(ind_test)
        python_solution = np.linalg.solve(all_A[ind_test], all_b[ind_test])
        py_solution.append(python_solution)
        eps.append(all_eps[0])
        solution = Seidel(all_A[ind_test], all_b[ind_test], all_eps[0])
        
        if not isinstance(solution, str):
            my_solution.append(solution[0])
            error.append(abs(solution[0] - python_solution))
            num_iter.append(solution[1])
        else:
            my_solution.append(solution)
            error.append('-')
            num_iter.append('-')

data = {'test': tests,
        'Точное решение': py_solution,
        'Допустимая погрешность решения': eps,
        'Полученное решение': my_solution,
        'Абсолютная погрешность решения': error,
        'Количество итераций': num_iter}

pd.DataFrame(data)
```

## Метод Гаусса (LU – разложение)

```python
def gaus_back(A, b):
    n = A.shape[0]
    x = np.zeros(n)
    m = np.concatenate((A, b.reshape(n,1)), axis=1)
    for k in range(n - 1, -1, -1):
        x[k] = (m[k, -1] - sum([m[k, j] * x[j] for j in range(k + 1, n)])) / m[k, k]
    return x


def LU(A, b):
    U = np.copy(A)
    our_b = np.copy(b)
    if U[0, 0] == 0:
        for i in range(1, A.shape[0]):
            if U[i, 0] != 0:
                U[[0, i]] = U[[i, 0]]
                our_b[0], our_b[i] = our_b[i], our_b[0]
                break
    L = np.eye(A.shape[0])
    def inv(matr):
        return np.linalg.inv(matr)
    for i in range(A.shape[0]):
        for j in range(i+1, A.shape[0]):
            x = U[j, i] / U[i,i]
            L[j, i] = x
            U[j] = U[j] - (x * U[i])
    inv_L = inv(L)
    Y = mult_matrix(inv_L, our_b)
    X = gaus_back(U, Y)
    return X
```

```python
tests = [] 
py_solution = []
my_solution = []
error = []

for ind_test in range(5):
    tests.append(ind_test)
    python_solution = np.linalg.solve(all_A[ind_test], all_b[ind_test])
    py_solution.append(python_solution)
    solution = LU(all_A[ind_test], all_b[ind_test]) # мое решение
    my_solution.append(solution)
    error.append(abs(solution - python_solution)) # погрешность
    
data = {'test': tests, 
        'Точное решение': py_solution,
        'Полученное решение': my_solution,
        'Абсолютная погрешность': error}

pd.DataFrame(data)
```

## QR - разложение

```python
from functools import reduce


def QR(A, b):
    shape = A.shape[0]
    Q_all = []
    Q_all.append(np.eye(shape)) 
    R_all = []
    R_all.append(np.copy(A)) 
    sub_R = np.copy(A)
    length = A.shape[0] 
    
    y = sub_R[:, 0]
    z = np.eye(length)[0]
    alf = vector_norm_2(y)
    ro = vector_norm_2(y -  alf*z)
    w = (y -  alf*z) / ro
    
    Q_all.append(matrix_sum(np.eye(length), -2 * mult_matrix(w.reshape(length,1), w.reshape(1,length))))
    R_all.append(mult_matrix(Q_all[1], R_all[0]))
    
    sub_R = R_all[1][1:,1:] 
    length -= 1 
    
    for i in range(1, shape-1):
        
        y = sub_R[:, 0]
        z = np.eye(length)[0] 
        alf = vector_norm_2(y)
        ro = vector_norm_2(y -  alf*z)
        w = (y -  alf*z) / ro
        
        new_Q = np.eye(shape)
        new_Q[i:, i:] = matrix_sum(np.eye(length), -2 * mult_matrix(w.reshape(length,1), w.reshape(1,length)))
        Q_all.append(new_Q)
        
        new_R = np.copy(R_all[i])
        new_R[i:, i:] = mult_matrix(new_Q[i:, i:], R_all[i][i:, i:])
        R_all.append(new_R)
        
        sub_R = new_R[i+1:, i+1:]
        length -= 1
        
    final_Q = reduce(lambda x, y: mult_matrix(x, y), Q_all) 
    final_R = R_all[-1]

    Q_T = matrix_T(final_Q)
    Y = mult_matrix(Q_T, b)
    X = gaus_back(final_R, Y)
    return X
```

```python
tests = [] 
py_solution = []
my_solution = []
error = []

for ind_test in range(5):
    tests.append(ind_test)
    python_solution = np.linalg.solve(all_A[ind_test], all_b[ind_test])
    py_solution.append(python_solution)
    solution = QR(all_A[ind_test], all_b[ind_test]) # мое решение
    my_solution.append(solution)
    error.append(abs(solution - python_solution)) # погрешность
    
data = {'test': tests, 
        'Точное решение': py_solution,
        'Полученное решение': my_solution,
        'Абсолютная погрешность': error}

pd.DataFrame(data)
```

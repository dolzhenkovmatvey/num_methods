```python
import math
import numpy as np
from tabulate import tabulate
```

```python
def sh(x, eps):
  x1 = 0
  k = 0
  while abs((x**(2*k+1))/(math.factorial(2*k+1)))/3 >= eps:
    x1+=(x**(2*k+1))/(math.factorial(2*k+1))
    k+=1
  return x1
```

```python
  def arctan(x, eps):
    if abs(x) >=1:
      x1 = np.sign(x)*(math.pi)/2
      k=0
      while abs(((-1)**k)*((x**((-2)*k-1))/(2*k+1))) >= eps:
        x1 -= ((-1)**k)*((x**((-2)*k-1))/(2*k+1))
        k+=1

    else:
      x1 = 0
      k=0
      while abs(((-1)**k)*(x**(2*k+1))/(2*k+1)) >= eps:
        x1 += ((-1)**k)*(x**(2*k+1))/(2*k+1)
        k+=1
    return x1
```

```python
def sq(x, eps):
  if x<0:
    return
  elif x >=1:
    p = x
    while abs(p - (1/2)*(p+x/p)) >= eps:
      p = (1/2)*(p+x/p)
  else:
    p = 1
    while abs(p - (1/2)*(p+x/p)) >= eps:
      p = (1/2)*(p+x/p)
  return p
```

```python
eps1 = 1e-6/3.3
eps2 = 1e-6/4.5
eps3 = 1e-6/3
eps = 1e-6
```

```python
x = np.linspace(0.01, 0.06, 11)
```

```python
table1 = [['x', 'Our shx', 'delta1', 'True shx', 'delta2']]
for i in x:
  table1.append([i, sh(2*i+0.45, eps1), eps1, math.sinh(2*i+0.45), abs(math.sinh(2*i+0.45)-sh(2*i+0.45, eps1))])
print(tabulate(table1))
```

```python
table2 = [['x', 'Our arctgx', 'delta1', 'True arctgx', 'delta2']]
for i in x:
  table2.append([i, arctan(6*i+1, eps2), eps2, math.atan(6*i+1),abs(math.atan(6*i+1)-arctan(6*i+1, eps2)) ])
print(tabulate(table2))
```

```python
table3 = [['x', 'Our sqrt', 'delta1', 'True sqrt', 'delta2']]
for i in x:
  table3.append([i, sq(i, eps3), eps3, math.sqrt(i), abs(math.sqrt(i)-sq(i, eps3))])
print(tabulate(table3))
```
